```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('responses.csv')
df
```

###### delete the column 'minsekRaw' as it is not useful in data analysis, more over this column data is in other two columns(min and sec)

```python
df1=df.drop(columns=['minSekRaw']) 
df1
```

###### look at the sec column, value should be between 0 and 59, if the value is greater than 59, calculate the seconds percent of 100.

```python
df1.loc[df1['sec']>59,['sec']] = df1.loc[df1['sec']>59,['sec']]*60/100
df1
```

###### when we consider accurate data, i.e; deleting NaN values

```python
df2=df1.dropna() 
df2
```

```python
df2['type'].value_counts()
```

```python
#df2.loc['min']+df2.loc['sec']

df2['sec'] = pd.to_numeric(df2['sec'],errors = 'coerce')

df2['min'] = pd.to_numeric(df2['min'],errors = 'coerce')
```

```python
df2.dtypes
```

```python
df2['time']=df2['min']*60+df2['sec']
```

```python
df2
```

```python
df2['time'].mean() # mean time of solving the problem
```

```python
#who was fastest

df2[df2['time']==df2['time'].min()]
```

###### So, Latin was fastest

```python
#whose mean time is low
greekMeanTime = df2[df2['type'] == 'Greek'].time.mean()
greekMeanTime
```

```python

latinMeanTime = df2[df2['type'] == 'Latin'].time.mean()
latinMeanTime
```

###### So, Greek mean time is low

```python
fig = plt.figure()
ax=fig.add_axes([1,1,0.5,0.5])
types=['Latin','Greek']
meanValues=[latinMeanTime,greekMeanTime]
ax.bar(types,meanValues,width=0.25)
```
